```python
import numpy as np
import pandas as pd
import geopandas as gpd
import gdutils.extract as et
```

```python
mggg_file = 'inputs/CT_precincts/CT_precincts.shp'
medsl_file = 'inputs/precinct_2018/precinct_2018.csv'
```

```python
mggg_gdf = gpd.read_file(mggg_file)
mggg_gdf.head()
```

```python
medsl_all_df = et.read_file(medsl_file).extract()
medsl_all_df.head()
```

```python
medsl_ct_df = et.ExtractTable(medsl_all_df, column='state', value='Connecticut').extract()
medsl_ct_df.head()
```

```python
medsl_pvt_df = medsl_ct_df.pivot_table(index='precinct', columns=['office', 'party'], values='votes')
medsl_pvt_df.head()
```

```python
medsl_df = medsl_pvt_df # make a copy
medsl_df.columns = medsl_pvt_df.columns = [''.join(col).strip() for col in medsl_pvt_df.columns.values]
medsl_df.head()
```
